# 시도별 교통사고 건수 데이터 시각화

## 데이터 불러오기

```python
import pandas as pd

doc = pd.read_csv('시도별_교통사고.csv', encoding='cp949')
doc.head()
```

## 필요한 컬럼만 불러오기

```python
doc = doc.drop([0], axis = 0)
```

```python
doc.rename(columns = {'시도별(1)': '시도별',
                      '월별(1)' : '월별'}, inplace = True)
```

```python
doc
```

## 연도와 월을 합치기

```python
doc = doc.melt(id_vars = ['시도별', '월별'])
doc
```

```python
doc['date'] = doc['variable'] + "/" + doc['월별']
doc.date = [x.replace('월', '') for x in list(doc.iloc[:,-1])]
```

```python
doc.drop(['월별', 'variable'], axis = 1, inplace = True)
```

```python
doc['value'] = doc['value'].replace('-', '0').astype(int)
```

```python
doc['누적교통사고건수'] = doc.groupby('시도별')['value'].cumsum()
```

```python
data = doc.pivot(index = '시도별', columns = 'date', values = '누적교통사고건수')
data = data.reindex(sorted(data.columns, key=pd.to_datetime), axis=1)
```

```python
data
```

## 파일 내보내기

```python
data.to_csv("./traffic_accident.csv")
```
